```python
# Importing our necessary libraries
import os
import pandas as pd
import numpy as np
```

```python
# Function to get all text files within one folder
def get_csv(path):
    csv_files=[]
    for file in os.listdir(path):
        if file.endswith(".txt"):
            csv_files.append(file)
    
    return csv_files
```
